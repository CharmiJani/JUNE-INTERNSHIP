```python
import matplotlib.pyplot as plt
ages=[2,5,70,40,30,45,43,40,44,60,7,13,57,18,90,77,32,21,40,23,56,43,21,34,81,81,81,81,81,81,83,88,90,12,43,22,12,58,25,3,33,67,32,30,90,50]

range=(0,100)
bins=20

plt.hist(ages , bins , range ,
color = 'purple', histtype = 'bar', edgecolor = 'white')

plt.xlabel('age')
plt.ylabel('No. of people')
plt.title('My histogram')
```

```python
import matplotlib.pyplot as plt
activities=['eat','sleep','work','play']
slices=[3,7,8,6]
colors=['r','b','g','y']
plt.pie(slices, labels = activities, colors = colors, startangle = 90, shadow = True, explode = (0,0.1,0,0), radius = 1.5, autopct = '%1.1f%%')
plt.show()
'''percentage = 3+7+8+6 = 24
eat = (3/24)*100 = 12.5%'''
```
